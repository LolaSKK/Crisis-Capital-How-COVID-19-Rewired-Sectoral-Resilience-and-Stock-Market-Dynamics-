```python
import logging
import time
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.firefox import GeckoDriverManager
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
logging.basicConfig(level=logging.CRITICAL)

options = webdriver.FirefoxOptions()
options.binary_location = "/share/fejermanlab/luqi/firefox/firefox"
options.add_argument("--headless")

service = Service(GeckoDriverManager(version="v0.34.0").install())

driver = webdriver.Firefox(service=service, options=options)

ordered_stock_names = []

for page_index in range(0, 300, 100):  # 0-100, 100-200, 200-300
    url = f"https://finance.yahoo.com/markets/stocks/most-active/?start={page_index}&count=100"
    driver.get(url)
    html = driver.execute_script('return document.body.innerHTML;')
    soup = BeautifulSoup(html, 'lxml')
    stock_names = [span.get_text() for span in soup.find_all('span', attrs={"class": re.compile("symbol")})]
    stock_names = stock_names[:100]
    ordered_stock_names.extend(stock_names)

url = "https://finance.yahoo.com/markets/stocks/most-active/?start=300&count=100"
driver.get(url)

html = driver.execute_script('return document.body.innerHTML;')
soup = BeautifulSoup(html, 'lxml')

stock_names = [span.get_text() for span in soup.find_all('span', attrs={"class": re.compile("symbol")})]
stock_names = stock_names[:52]

ordered_stock_names.extend(stock_names)

driver.quit()

print(len(ordered_stock_names))
```

```python
ordered_stock_names = [stock.strip() for stock in ordered_stock_names]

print(ordered_stock_names)
```

```python
import yfinance as yf
data = yf.download(ordered_stock_names, start="2019-01-01", end="2025-03-10", group_by='tickers')
```

```python
ticker = "AAPL"

ticker_object = yf.Ticker(ticker)

temp = pd.DataFrame.from_dict(ticker_object.info, orient="index")
temp.reset_index(inplace=True)
temp.columns = ["Attribute", "Recent"]

pd.set_option('display.max_rows', None)

print(temp)
```

```python
tickers_data = {}

for ticker in ordered_stock_names:
    ticker_object = yf.Ticker(ticker)
    industry = ticker_object.info.get("industry", "N/A")
    tickers_data[ticker] = industry
    time.sleep(1)

print(tickers_data)
```

```python
df = pd.DataFrame(list(tickers_data.items()), columns=['Ticker', 'Industry'])

print(df)
```

```python
df.to_csv('tickers_industry_data.csv', index=False)
```
